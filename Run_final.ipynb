```python
#checklist
from checklist.editor import Editor
from checklist.test_suite import TestSuite
from checklist.test_types import MFT   # Minimum Functionality Test
from checklist.expect import Expect
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
from checklist.editor import Editor
from checklist.perturb import Perturb
from transformers import pipeline
from checklist.test_types import MFT, INV, DIR

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
model = AutoModelForSequenceClassification.from_pretrained(model_checkpoint)

text_classification_pipeline = pipeline("text-classification", model=model_checkpoint)


# Predict function using the model and tokenizer

def pred_and_conf2(data):
    # Assuming `data` is a list of texts
    results = text_classification_pipeline(data, return_all_scores=True)

    # Process the results to extract predictions and confidence scores
    preds = [max(result, key=lambda x: x['score'])['label'] for result in results]
    confs = [max(result, key=lambda x: x['score'])['score'] for result in results]

    return preds, confs

import re

def predict(examples):
    predictions = []
    confidences = []

    # You need to provide a logical pairing for your examples here.
    # This is a placeholder loop that assumes each example is a hypothesis
    # and the premise is a fixed sentence. This is just for demonstration purposes.
    premise = "Some men are walking in the park."
    for hypothesis in examples:
        # Tokenize the premise and hypothesis
        print(str(premise)+":"+str(hypothesis))
        inputs = tokenizer.encode_plus(str(premise), str(hypothesis), return_tensors='pt', padding=True, truncation=True)

        # Get the logits from the model
        with torch.no_grad():
            outputs = model(**inputs)

        # Convert logits to probabilities using softmax
        probs = torch.softmax(outputs.logits, dim=-1)

        # Get the predicted label index with the highest probability
        predicted_label_idx = torch.argmax(probs, dim=-1).item()
        # Convert the index to the corresponding label
        label_map = {0: 'entailment', 1: 'neutral', 2: 'contradiction'}
        predicted_label = label_map[predicted_label_idx]

        predictions.append(predicted_label)
        #confidences.append(probs.squeeze().tolist())

        # Select the confidence for the predicted class only
        predicted_confidence = probs[0][predicted_label_idx].item()  # This should be a float, not a list
        confidences.append(predicted_confidence)

    return predictions, confidences
```

```python
from transformers import AutoModelForQuestionAnswering, AutoTokenizer
from datasets import load_dataset
from torch.utils.data import DataLoader
import torch

# Load the model and tokenizer
model_checkpoint = "trained_model_qa_64/trained_model_qa_64/checkpoint-4000"
model = AutoModelForQuestionAnswering.from_pretrained(model_checkpoint)
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)

# Load the SQuAD dataset
dataset = load_dataset("squad", split="validation")

# Tokenize the dataset
def tokenize_function(examples):
    return tokenizer(
        examples['question'], examples['context'],
        max_length=384, truncation=True, padding='max_length', return_tensors='pt'
    )

# Apply tokenization to the dataset
eval_dataset = dataset.map(tokenize_function, batched=True)

# Set the dataset format to only include the necessary columns
eval_dataset
eval_dataset.set_format(type='torch', columns=['input_ids', 'attention_mask', 'id'])

# Create a DataLoader for our dataset
eval_dataloader = DataLoader(eval_dataset, batch_size=64)

# Get model predictions
model.eval()  # Set the model to evaluation mode
predictions = []

# Loop through the DataLoader output
for batch in eval_dataloader:
    # Send the batch to the same device as the model
    batch = {k: v.to(model.device) for k, v in batch.items() if isinstance(v, torch.Tensor)}
    input_ids = batch['input_ids']
    with torch.no_grad():
        outputs = model(**batch)

    # Extract the start and end logits
    answer_start_scores = outputs.start_logits
    answer_end_scores = outputs.end_logits

    # Get the most probable start and end indices for the answer
    answer_starts = torch.argmax(answer_start_scores, dim=1)
    answer_ends = torch.argmax(answer_end_scores, dim=1) + 1

    # Convert the indices to lists for further processing
    #predictions.extend(zip(answer_starts.tolist(), answer_ends.tolist()))
    # Extract the answer tokens from the input IDs
    for start, end, input_id in zip(answer_starts, answer_ends, input_ids):
        # Add 1 to end index to include the end token
        answer = tokenizer.convert_tokens_to_string(
            tokenizer.convert_ids_to_tokens(input_id[start:end+1], skip_special_tokens=True)
        )
        predictions.append(answer)
```

```python
from collections import Counter

# Define a function to compute the F1 score for individual predictions
def compute_f1(a_gold, a_pred):
    gold_toks = a_gold.split()
    pred_toks = a_pred.split()
    common = Counter(gold_toks) & Counter(pred_toks)
    num_same = sum(common.values())
    if len(gold_toks) == 0 or len(pred_toks) == 0:
        # If either is no-answer, then F1 is 1 if they agree, 0 otherwise
        return int(gold_toks == pred_toks)
    if num_same == 0:
        return 0
    precision = 1.0 * num_same / len(pred_toks)
    recall = 1.0 * num_same / len(gold_toks)
    f1 = (2 * precision * recall) / (precision + recall)
    return f1

# Helper function to categorize errors
def categorize_error_type(gold_answers, prediction, context):
    # If prediction is empty and there is an answer or vice versa
    if not prediction.strip() and gold_answers:
        return "No Answer"
    if prediction.strip() and not gold_answers:
        return "Spurious Answer"

    # If prediction is not empty and is not found in the gold answers
    if prediction.strip() and not any(prediction in answer for answer in gold_answers):
        # Check if the predicted answer is in the context
        if prediction in context:
            return "Wrong Location"
        else:
            return "Incorrect Answer"

    # If prediction is partially correct
    if any(prediction.strip() in answer for answer in gold_answers):
        return "Correct Answer with Wrong Words"

    return "Other"

# Helper function to categorize errors
def categorize_error(predicted, actuals):
    # Normalize answers for comparison (e.g., lowercase)
    predicted_norm = predicted.lower().strip()
    actuals_norm = [answer.lower().strip() for answer in actuals]

    if predicted_norm in actuals_norm:
        return 'Exact Match'
    elif any(predicted_norm in answer for answer in actuals_norm):
        return 'Partial Match'
    elif predicted_norm and all(predicted_norm not in answer for answer in actuals_norm):
        return 'No Match'
    else:
        return 'Invalid Prediction'
```

```python
#aquad dataset
#單詞檢測
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Assuming you have already loaded your model and tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
model = AutoModelForSequenceClassification.from_pretrained(model_checkpoint)
model.eval()  # Put the model into evaluation mode

# Assuming you have your SQuAD dataset loaded and in the form of a list of dicts
squad_dataset = [{'context': "The Normans (Norman: Nourmands; French: Normands; Latin: Normanni) were the people who in the 10th and 11th centuries gave their name to Normandy, a region in France.",
                  'question': "In what country is Normandy located?",
                  'answers': ["France"]},
                 # ... more data
                ]

# Validate the model on the SQuAD dataset
for example in squad_dataset:
    context = example['context']
    question = example['question']
    answers = example['answers']

    # You would have to decide how to use the context and answer(s) to create a premise and hypothesis
    # Here's one simple way: use the first answer and the surrounding context as the premise
    answer = answers[0]  # Taking the first answer
    premise = context[context.find(answer):]  # The context starting from the answer
    hypothesis = question

    # Tokenize the inputs
    inputs = tokenizer.encode_plus(premise, hypothesis, return_tensors='pt', padding=True, truncation=True)

    # Get the model predictions
    with torch.no_grad():  # Disabling gradient calculation
        outputs = model(**inputs)

    logits = outputs.logits
    predicted_label_index = logits.argmax(-1).item()

    # Convert the predicted index to a label
    label_map = {0: "entailment", 1: "neutral", 2: "contradiction"}
    predicted_label = label_map[predicted_label_index]

    print(f"Question: {question}")
    print(f"Context: {context}")
    print(f"Predicted NLI label: {predicted_label}")
```
